## *Prediction using unsupervised machine learning*

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
# Load Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Target labels
```

```python
# Initialize an empty list to store the within-cluster sum of squares (WCSS) for different k values
wcss = []
```

```python
# Iterate through different values of k (number of clusters)
for i in range(1, 11):
    # Initialize KMeans object with the current value of k
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    # Fit KMeans model to the data
    kmeans.fit(X)
    # Append the WCSS value to the list
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the elbow method graph to find the optimal number of clusters
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

**Based on the elbow method graph, choose the optimal number of clusters.
In this case, it appears that the elbow point is around k=3.**

```python
# Applying KMeans with the chosen number of clusters
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
# Fit KMeans model to the data
kmeans.fit(X)
```

```python
# Get the cluster centers
centroids = kmeans.cluster_centers_
print("Cluster centers:")
print(centroids)
```

```python
# Get the labels assigned to each data point
labels = kmeans.labels_
print("\nCluster labels for each data point:")
print(labels)
```

```python
# Visualizing the clusters (considering only the first two features for simplicity)
plt.scatter(X[labels == 0, 0], X[labels == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[labels == 2, 0], X[labels == 2, 1], s=100, c='green', label='Cluster 3')
```

```python
# Plotting the centroids of the clusters
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='yellow', label='Centroids')
plt.title('Clusters of Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```
